```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/media/data_4T/bran/WMH_dataset/raw/training/center_specific_flair_png/val/images/GE3T_132_53.png', -1)
swin_45 = cv2.imread('/home/margaryta/my_training/visual/132_swin_53.png', -1)
unet_45 = cv2.imread('/home/margaryta/my_training/visual/132_unet_53.png', -1)
gt = cv2.imread('/media/data_4T/bran/WMH_dataset/raw/training/center_specific_flair_png/val/masks/GE3T_132_53.png', -1)

unet_45 = unet_45[...,::-1]
swin_45 = swin_45[...,::-1]
gt = gt[...,::-1]

img = np.rot90(img, k=3)
swin_45 = np.rot90(swin_45, k=3)
unet_45 = np.rot90(unet_45, k=3)

gt = np.rot90(gt, k=3)
gt = 255-gt
```

```python
x1 = 55
x2 = 200
y1 = 50
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,9))

ax1.imshow(img[y1:, x1:x2], cmap='gray')
ax4.imshow(swin_45[y1:,x1:x2,:])
ax3.imshow(unet_45[y1:,x1:x2,:])
ax2.imshow(gt[y1:,x1:x2])
'''
ax1.set_title('FLAIR', fontsize=20)
ax4.set_title('Segmentation \n Swin Transformer', fontsize=20)
ax3.set_title('Segmentation \n UNet', fontsize=20)
ax2.set_title('Ground Truth', fontsize=20)
'''
ax1.set_ylabel('GE3T patient 132 \n slice 53', fontsize=20)
ax1.tick_params(labelleft=False)  

ax1.xaxis.set_visible(False)
ax1.yaxis.set_visible(True)
ax2.xaxis.set_visible(False)
ax2.yaxis.set_visible(False)
ax3.xaxis.set_visible(False)
ax3.yaxis.set_visible(False)
ax4.xaxis.set_visible(False)
ax4.yaxis.set_visible(False)

fig.patch.set_facecolor('white')

plt.savefig('132_53.png', bbox_inches='tight')
```
